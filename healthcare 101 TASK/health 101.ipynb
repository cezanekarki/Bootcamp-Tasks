```
%sql
INSERT INTO target_table_h101(
    Abacus_Record_ID,
    Abacus_Member_ID,
    Member_ID,
    Subscriber_ID,
    Member_First_Name,
    Member_Last_Name,
    Member_Middle_Name,
    Member_Prefix_Name,
    Member_Suffix_Name,
    Member_Gender,
    Member_Date_of_Birth,
    Member_Relationship_Code,
    Member_Relationship_Description,
    Member_Person_Code,
    Member_Address_Line_1,
    Member_Address_Line_2,
    Member_City,
    Member_State,
    Member_County,
    Member_Postal_Code,
    Member_Country,
    Member_Home_Phone,
    Member_Work_Phone,
    Member_Mobile_Phone,
    Member_Email,
    Member_Is_Deceased,
    Member_Date_of_Death,
    Member_Deceased_Reason,
    Enrollment_Group_ID,
    Enrollment_Group_Name,
    Enrollment_SubGroup_ID,
    Enrollment_SubGroup_Name,
    Enrollment_Coverage_Code,
    Enrollment_Coverage_Description,
    Enrollment_Plan_ID,
    Enrollment_Plan_Name,
    Enrollment_Plan_Coverage,
    Enrollment_Medical_Effective_Date,
    Enrollment_Medical_Termination_Date,
    Enrollment_Dental_Effective_Date,
    Enrollment_Dental_Termination_Date,
    Enrollment_Vision_Effective_Date,
    Enrollment_Vision_Termination_Date,
    Enrollment_Vendor_Name,
    Souce_File_Name,
    File_Ingestion_Date
)
SELECT
    row_number() OVER ( ORDER BY(1)) AS Abacus_Record_ID,
    concat(
        demo.Employee_ID,
        row_number() OVER (ORDER BY(null)),
        case
            when demo.GENDER = 1 then 'Male'
            when demo.GENDER = 2 then 'Unknown'
            when demo.GENDER = 0 then 'Female'
        END,
        REPLACE(demo.DOB,"-"," ")
    ) AS Abacus_Member_ID,
    demo.MEMBER_ID as Member_ID,
    demo.EMPLOYEE_ID as Subscriber_ID,
    demo.FIRST_NAME as Member_First_Name,
    demo.LAST_NAME as Member_Last_Name,
    demo.MIDDLE_NAME as Member_Middle_Name,
    CASE
        WHEN gen.Rollup_Description = "Male" THEN "Mr."
        WHEN gen.Rollup_Description = "Female" THEN "Mrs."
        ELSE "Unknown"
    END AS Member_Prefix_Name,
    NULL AS Member_Suffix_Name,
    gen.Rollup_Description AS Member_Gender,
    demo.DOB AS Member_Date_of_Birth,
    demo.Relationship AS Member_Relationship_Code,
    rel.Rollup_Description AS Member_Relationship_Description,
    demo.PERSON_CODE AS Member_Person_Code,
    demo.ADDRESS_1 as Member_Address_Line_1,
    demo.ADDRESS_1 as Member_Address_Line_2,
    demo.CITY as Member_City,
    demo.STATE as Member_State,
    demo.COUNTY as Member_County,
    demo.ZIP as Member_Postal_Code,
    "US" as Member_Country,
    NULL as Member_Home_Phone,
    NULL as Member_Work_Phone,
    NULL as Member_Mobile_Phone,
    NULL as Member_Email,
    NULL as Member_Is_Deceased,
    NULL as Member_Date_of_Death,
    NULL as Member_Deceased_Reason,
    enrol.GROUP_ID as Enrollment_Group_ID,
    grp.GROUP_NAME as Enrollment_Group_Name,
    NULL as Enrollment_SubGroup_ID,
    NULL as Enrollment_SubGroup_Name,
    cvg.Coverage_Description as Enrollment_Coverage_Description,
    enrol.PLAN_ID as Enrollment_Plan_ID,
    pln.`Plan Name` as Enrollment_Plan_Name,
    pln.BENEFIT_TYPE as Enrollment_Plan_Coverage,
    enrol.EFFECTIVE_DATE AS Enrollment_Medical_Effective_Date,
    CASE 
        WHEN Pln.benefit_type LIKE '%Medical%' THEN Pln.EFFECTIVE_DATE
        ELSE NULL
    END AS Enrollment_Medical_Effective_Date,
    CASE 
        WHEN Pln.benefit_type LIKE '%Medical%' THEN Pln.TERMINATION_DATE
        ELSE NULL
    END AS Enrollment_Medical_Termination_Date,
    CASE 
        WHEN Pln.benefit_type LIKE '%Dental%' THEN Pln.EFFECTIVE_DATE
        ELSE NULL
    END AS Enrollment_Dental_Effective_Date,
    CASE 
        WHEN Pln.benefit_type LIKE '%Dental%' THEN Pln.TERMINATION_DATE
        ELSE NULL
    END AS Enrollment_Dental_Termination_Date,
    CASE 
        WHEN Pln.benefit_type LIKE '%Vision%' THEN Pln.EFFECTIVE_DATE
        ELSE NULL
    END AS Enrollment_Vision_Effective_Date,
    CASE 
        WHEN Pln.benefit_type LIKE '%Vision%' THEN Pln.TERMINATION_DATE
        ELSE NULL
    END AS Enrollment_Vision_Termination_Date,
    enrol.VENDOR as Enrollment_Vendor_Name,
    'Eligibility Mock Data' as Souce_File_Name,
    CURRENT_TIMESTAMP as File_Ingestion_Date

FROM
    demographics demo
JOIN 
    enrollment enrol ON demo.MEMBER_ID = enrol.MEMBER_ID
JOIN
    gender_cw gen on demo.GENDER = gen.Code
JOIN 
    relationship_cw rel on demo.RELATIONSHIP = rel.Rollup_Code
JOIN 
    group_cw grp on enrol.GROUP_ID = grp.GROUP_ID
JOIN
    plan_cw pln on enrol.PLAN_ID = pln.PLAN_ID 
JOIN 
    coverage_cw cvg on enrol.COVERAGE_TYPE = cvg.Coverage_ID;
```
