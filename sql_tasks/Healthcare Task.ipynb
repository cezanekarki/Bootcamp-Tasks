```
from pyspark.sql import SparkSession, functions as F
 
spark=SparkSession.builder.getOrCreate()
 
path = '/FileStore/tables/Eligibility_Mock_Data___US_Healthcare_Bootcamp-5.xlsx'
dataFormat = "com.crealytics.spark.excel"
 
enrollment_sheet_name="Enrollment Mock Data"
demographics_sheet_name="Demographics Mock Data"
 
enrollment_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{enrollment_sheet_name}'!").load(path)
demographics_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{demographics_sheet_name}'!").load(path)

gender_crosswalk="Gender Crosswalk"
relationship_crosswalk="Relationship Crosswalk"
group_crosswalk = "Group Crosswalk"
coverage_crosswalk = "Coverage Crosswalk"
plan_crosswalk = "Plan Crosswalk"
 
gender_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{gender_crosswalk}'!").load(path)
relationship_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{relationship_crosswalk}'!").load(path)
group_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{group_crosswalk}'!").load(path)
coverage_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{coverage_crosswalk}'!").load(path)
plan_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{plan_crosswalk}'!").load(path)
```

```
%sql
CREATE TABLE Abacus_Record (
    Abacus_Record_ID VARCHAR(255),
    Abacus_Member_ID VARCHAR(255),
    Member_ID VARCHAR(255),
    Subscriber_ID VARCHAR(255),
    Member_First_Name VARCHAR(255),
    Member_Last_Name VARCHAR(255),
    Member_Middle_Name VARCHAR(255),
    Member_Prefix_Name VARCHAR(255),
    Member_Suffix_Name VARCHAR(255),
    Member_Gender VARCHAR(255),
    Member_Date_of_Birth DATE,
    Member_Relationship_Code VARCHAR(255),
    Member_Relationship_Description VARCHAR(255),
    Member_Person_Code INTEGER,
    Member_Address_Line_1 VARCHAR(255),
    Member_Address_Line_2 VARCHAR(255),
    Member_City VARCHAR(255),
    Member_State VARCHAR(255),
    Member_County VARCHAR(255),
    Member_Postal_Code VARCHAR(255),
    Member_Country VARCHAR(255),
    Member_Home_Phone INTEGER,
    Member_Work_Phone INTEGER,
    Member_Mobile_Phone INTEGER,
    Member_Email VARCHAR(255),
    Member_Is_Deceased CHAR(1),
    Member_Date_of_Death DATE,
    Member_Deceased_Reason VARCHAR(255),
    Enrollment_Group_ID VARCHAR(255),
    Enrollment_Group_Name VARCHAR(255),
    Enrollment_SubGroup_ID VARCHAR(255),
    Enrollment_SubGroup_Name VARCHAR(255),
    Enrollment_Coverage_Code VARCHAR(255),
    Enrollment_Coverage_Description VARCHAR(255),
    Enrollment_Plan_ID VARCHAR(255),
    Enrollment_Plan_Name VARCHAR(255),
    Enrollment_Plan_Coverage VARCHAR(255),
    Enrollment_Medical_Effective_Date DATE,
    Enrollment_Medical_Termination_Date DATE,
    Enrollment_Dental_Effective_Date DATE,
    Enrollment_Dental_Termination_Date DATE,
    Enrollment_Vision_Effective_Date DATE,
    Enrollment_Vision_Termination_Date DATE,
    Enrollment_Vendor_Name VARCHAR(255),
    Souce_File_Name VARCHAR(255),
    File_Ingestion_Date DATE
);
```

```
%sql
INSERT INTO Abacus_Record(
    Abacus_Record_ID,
    Abacus_Member_ID,
    Member_ID,
    Subscriber_ID,
    Member_First_Name,
    Member_Last_Name,
    Member_Middle_Name,
    Member_Prefix_Name,
    Member_Suffix_Name,
    Member_Gender,
    Member_Date_of_Birth,
    Member_Relationship_Code,
    Member_Relationship_Description,
    Member_Person_Code,
    Member_Address_Line_1,
    Member_Address_Line_2,
    Member_City,
    Member_State,
    Member_County,
    Member_Postal_Code,
    Member_Country,
    Member_Home_Phone,
    Member_Work_Phone,
    Member_Mobile_Phone,
    Member_Email,
    Member_Is_Deceased,
    Member_Date_of_Death,
    Member_Deceased_Reason,
    Enrollment_Group_ID,
    Enrollment_Group_Name,
    Enrollment_SubGroup_ID,
    Enrollment_SubGroup_Name,
    Enrollment_Coverage_Code,
    Enrollment_Coverage_Description,
    Enrollment_Plan_ID,
    Enrollment_Plan_Name,
    Enrollment_Plan_Coverage,
    Enrollment_Medical_Effective_Date,
    Enrollment_Medical_Termination_Date,
    Enrollment_Dental_Effective_Date,
    Enrollment_Dental_Termination_Date,
    Enrollment_Vision_Effective_Date,
    Enrollment_Vision_Termination_Date,
    Enrollment_Vendor_Name,
    Souce_File_Name,
    File_Ingestion_Date
)
SELECT
    ROW_NUMBER() OVER (order by 1) AS Abacus_Record_ID,
    CONCAT(
    demo.EMPLOYEE_ID,
    ROW_NUMBER() OVER (ORDER BY (SELECT NULL)),
    CASE 
        WHEN demo.GENDER = 1 THEN '1'
        WHEN demo.GENDER = 0 THEN '0'
        WHEN demo.GENDER = 2 THEN '2'
    END,
    REPLACE(demo.DOB, '-', '')
) AS Abacus_Member_ID,
    demo.MEMBER_ID as Member_ID,
    demo.EMPLOYEE_ID as Subscriber_ID,
    demo.FIRST_NAME as Member_First_Name,
    demo.LAST_NAME as Member_Last_Name,
    demo.MIDDLE_NAME as Member_Middle_Name,
    CASE 
        WHEN gen.Rollup_Description = 'Male' THEN 'Mr.'
        WHEN gen.Rollup_Description = 'Female' THEN 'Ms.'
        ELSE 'Unknown'
    END as Member_Prefix_Name,
    NULL as Member_Suffix_Name,
    gen.Rollup_Description as Member_Gender,
    demo.DOB as Member_Date_of_Birth,
    demo.RELATIONSHIP as Member_Relationship_Code,
    rel.Rollup_Description as Member_Relationship_Description,
    demo.PERSON_CODE as Member_Person_Code,
    demo.ADDRESS_1 as Member_Address_Line_1,
    demo.ADDRESS_1 as Member_Address_Line_2,
    demo.CITY as Member_City,
    demo.STATE as Member_State,
    demo.COUNTY as Member_County,
    demo.ZIP as Member_Postal_Code,
    "US" as Member_Country,
    NULL as Member_Home_Phone,
    NULL as Member_Work_Phone,
    NULL as Member_Mobile_Phone,
    NULL as Member_Email,
    NULL as Member_Is_Deceased,
    NULL as Member_Date_of_Death,
    NULL as Member_Deceased_Reason,
    enro.GROUP_ID as Enrollment_Group_ID,
    grp.GROUP_NAME as Enrollment_Group_Name,
    NULL as Enrollment_SubGroup_ID,
    NULL as Enrollment_SubGroup_Name,
    enro.COVERAGE_TYPE as Enrollment_Coverage_Code,
    cvg.Coverage_Description as Enrollment_Coverage_Description,
    enro.PLAN_ID as Enrollment_Plan_ID,
    pln.`Plan Name` as Enrollment_Plan_Name,
    pln.BENEFIT_TYPE as Enrollment_Plan_Coverage,
    CASE 
    WHEN Pln.benefit_type LIKE '%Medical%' THEN Pln.EFFECTIVE_DATE
    ELSE NULL
END AS Enrollment_Medical_Effective_Date,

CASE 
    WHEN Pln.benefit_type LIKE '%Medical%' THEN Pln.TERMINATION_DATE
    ELSE NULL
END AS Enrollment_Medical_Termination_Date,

CASE 
    WHEN Pln.benefit_type LIKE '%Dental%' THEN Pln.EFFECTIVE_DATE
    ELSE NULL
END AS Enrollment_Dental_Effective_Date,

CASE 
    WHEN Pln.benefit_type LIKE '%Dental%' THEN Pln.TERMINATION_DATE
    ELSE NULL
END AS Enrollment_Dental_Termination_Date,

CASE 
    WHEN Pln.benefit_type LIKE '%Vision%' THEN Pln.EFFECTIVE_DATE
    ELSE NULL
END AS Enrollment_Vision_Effective_Date,

CASE 
    WHEN Pln.benefit_type LIKE '%Vision%' THEN Pln.TERMINATION_DATE
    ELSE NULL
END AS Enrollment_Vision_Termination_Date,

    enro.VENDOR as Enrollment_Vendor_Name,
    'Eligibility Mock Data' as Souce_File_Name,
    CURRENT_TIMESTAMP as File_Ingestion_Date
FROM
    demographics demo
JOIN 
    enrollment enro ON demo.MEMBER_ID = enro.MEMBER_ID
JOIN
    gender gen on demo.GENDER = gen.Code
JOIN 
    relationship rel on demo.RELATIONSHIP = rel.Rollup_Code
JOIN 
    group grp on enro.GROUP_ID = grp.GROUP_ID
JOIN
    plan pln on enro.PLAN_ID = pln.PLAN_ID 
JOIN 
    coverage cvg on enro.COVERAGE_TYPE = cvg.Coverage_ID
```
