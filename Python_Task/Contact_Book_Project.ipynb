#**Project: Contact Book Application**

```python
# @title Importing the required library
from IPython.display import display, clear_output
import ipywidgets as ipw
import json
```

```python
# @title Initializing the empty list
contact_list = []
```

```python
# @title Add contact Functionality
def add_contact(*args):
    return {'name':args[0],'phone': args[1], 'email': args[2]}
def on_submit_clicked(b):
    # Call the function to add the contact
    contact = add_contact(name_text.value, phone_text.value, email_text.value)
    contact_list.append(contact)
    # Optionally clear the input fields after adding
    name_text.value = ''
    phone_text.value = 0
    email_text.value = ''
    clear_output(wait=True)
    main()
def display_add(b):
  clear_output(wait=True)
  global name_text
  global phone_text
  global email_text
  name_text = ipw.Text(description="Name:")
  phone_text = ipw.IntText(description="Phone:")
  email_text = ipw.Text(description="Email:")
  submit_button = ipw.Button(description="Add Contact")
  submit_button.on_click(on_submit_clicked)
  display(name_text, phone_text, email_text, submit_button)

```

```python
# @title View All Contact Functionality
def display_all_contact(b):
  json_contacts = json.dumps(contact_list, indent=4)
  clear_output(wait=True)
  print(json_contacts)
  main()
```

```python
# @title Search Contact Functionality
def on_search_click(name_text):
    search_query = name_text.value
    for contact in contact_list:
        if contact['name'].lower() == search_query.lower():
          search_json = json.dumps(contact, indent=4)
          clear_output(wait=True)
          print(search_json)
          main()
          return
    print("Contact not found.")
def display_search_field(b):
  name_text = ipw.Text(description="Name:")
  search_button = ipw.Button(description="Search By Name")
  search_button.on_click(lambda b: on_search_click(name_text))
  display(name_text, search_button)
```

```python
# @title Delete Contact Functionality
def on_delete_click(name_text):
    global contact_list
    to_delete_item = name_text.value
    contact_list = [contact for contact in contact_list if contact['name'].lower() != to_delete_item.lower()]
    json_contacts = json.dumps(contact_list, indent=4)
    clear_output(wait=True)
    print(f"Updated contact list: {json_contacts}")
    main()

def display_delete_field(b):
  name_text = ipw.Text(description="Name:")

  # Create a button widget for submitting the form
  delete_button = ipw.Button(description="Delete By Name")
  delete_button.on_click(lambda b: on_delete_click(name_text))
  display(name_text, delete_button)
```

```python
# @title Exit Functionality
import ipywidgets as widgets
from IPython.display import clear_output, display

def quit_button_click(b):
    clear_output(wait=True)
    print("The output has been cleared!")


```

```python
# @title Main Menu
from google.colab import widgets
from IPython.display import display, clear_output
import ipywidgets as ipw
def main():
  add_contact = ipw.Button(description="Add Contact")
  view_contact = ipw.Button(description="View All Contact")
  search_contact = ipw.Button(description="Search Contact")
  delete_contact = ipw.Button(description="Delete Contact")
  quit_button = ipw.Button(description="Quit")
  display(add_contact,view_contact,search_contact,delete_contact,quit_button)
  add_contact.on_click(display_add)
  view_contact.on_click(display_all_contact)
  search_contact.on_click(display_search_field)
  delete_contact.on_click(display_delete_field)
  quit_button.on_click(quit_button_click)



```

```python
# @title Calling the Main Method
main()
```
