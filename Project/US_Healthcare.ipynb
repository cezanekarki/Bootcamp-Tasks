```
%sql
SELECT 
      concat(d.employee_id," ",e.member_id," ",d.gender," ",d.DOB)as Abacus_Mid,
      e.member_id,
      d.employee_id as suscriber_id,
      d.first_name,
      d.middle_name,
      d.last_name,
    --case when d.gender = 0 then "Mr" when d.gender = 1 and r.Rollup_code in ('s','m','g2') then 'Mrs' else 'MS' end as Prefix
    "null" as Prefix,
    "null" as Sufix,
    case when d.gender = 0 then "Male" when d.gender  = 1 then "Female"  else "Unknown"end as Gender,
    date(CONCAT(EXTRACT(MONTH FROM d.DOB), '-', EXTRACT(DAY FROM d.DOB), '-', EXTRACT(YEAR FROM d.DOB))) AS Date_of_birth,
    d.relationship as relationship_code,
    r.Rollup_description as Relationship_desc,
    d.person_code as member_person_code,
    d.ADDRESS_1 as  Address_line1,
    d.ADDRESS_2 as Address_line_2,
    d.city as City,
    d.STATE,d.COUNTY,
    "null" as Member_Home_phone,
    "null" as Member_Work_phone,
    "null" as Member_mobile_phone,
    "null" as Member_Email,
    "null" AS  Member_is_Deceased,
    "null" as Date_of_death,
    "null" as member_death_reason,
    d.ZIP AS Posatl_code,
    "USA" as Member_country,
    e.Group_id,
    grp.GROUP_NAME,
    "null" as Sub_group_id,
    "null" as Sub_group_Name,
    e.Coverage_Type as Coverage_Code,
    c.Coverage_Description,
    e.Plan_id,
    p.plan_Name,
    p.BENEFIT_TYPE as Plan_Coverage,
    --e.EFFECTIVE_DATE,e.TERMINATION_DATE
    CASE 
    WHEN BENEFIT_TYPE LIKE '%Medical%' THEN p.EFFECTIVE_DATE 
    ELSE NULL
END AS Medical_effective_date,

CASE 
    WHEN BENEFIT_TYPE LIKE '%Medical%' THEN p.TERMINATION_DATE 
    ELSE NULL
END AS Medical_terminative_date,

CASE 
    WHEN BENEFIT_TYPE LIKE '%Dental%' THEN p.EFFECTIVE_DATE 
    ELSE NULL
END AS Dental_effective_date,

CASE 
    WHEN BENEFIT_TYPE LIKE '%Dental%' THEN p.TERMINATION_DATE 
    ELSE NULL
END AS Dental_terminative_date,
CASE 
    WHEN BENEFIT_TYPE LIKE '%vISION%' THEN p.EFFECTIVE_DATE 
    ELSE NULL
END AS Vision_effective_date,
CASE 
    WHEN BENEFIT_TYPE LIKE '%Vision%' THEN p.TERMINATION_DATE 
    ELSE NULL
END AS Vision_termination_date,
e.VENDOR as Vendor_name,
'Member_Enrollmnt_data' as Source_File,
current_date() as File_ingestion_date






FROM 
    demographic d 
LEFT JOIN 
    enrollment e ON d.MEMBER_ID = e.MEMBER_ID 
LEFT JOIN  
    Gender g ON d.gender = g.code 
LEFT JOIN 
    relationship r ON d.relationship = r.Rollup_code 
LEFT JOIN 
    Group grp ON e.group_id = grp.Group_id 
LEFT JOIN  
    coverage c ON e.coverage_type = c.coverage_id 
LEFT JOIN 
    plan1 p ON e.plan_id = p.plan_id
```
