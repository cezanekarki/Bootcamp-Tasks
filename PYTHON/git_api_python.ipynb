```
import requests

class Session:
    def __init__(self):
        # Initialize any session-specific settings here
        self.session = requests.Session()

    def make_get_request(self, url):
        # Use the session's post method to make a POST request
        response = self.session.get(url)
        return response

# Instantiate the Session class
my_session = Session()

# Specify the URL where you want to make the POST request
get_url = "https://gitlab.com/gitlab-org/gitlab-docs"

# Make the POST request using the session
response = my_session.make_get_request(get_url)

# Print the response content
print(response.text)
```

```
import requests
import json
from datetime import datetime 

# Replace 'your_private_token' with your GitLab personal access token
private_token = 'glpat-8KFNcmr2yViNfQYhNa6j'
# Replace 'gitlab-org%2Fgitlab-docs' with your project path
project_path = 'gitlab-org%2Fgitlab-docs'
project_id = '1794617'  # Replace with your actual project ID

# API endpoint to get contributors
contributors_api_url = f'https://gitlab.com/api/v4/projects/{project_id}/repository/contributors'
# API endpoint to get repository tree
tree_api_url = f'https://gitlab.com/api/v4/projects/{project_path}/repository/tree'
# Define the GitLab API endpoint for project details
project_details_url = f'https://gitlab.com/api/v4/projects/{project_id}'

# Parameters for the API requests
params = {'private_token': private_token, 'id': project_id,'recursive': True,'per_page': 100}

try:

    # Make the API request to get contributors
    contributors_response = requests.get(contributors_api_url, params=params)
    contributors_data = contributors_response.json()

    # Make the API request to get repository tree
    tree_response = requests.get(tree_api_url, params=params)
    tree_data = tree_response.json()

    # Make the API request to get project details
    project_response = requests.get(project_details_url, params=params)
    project_data = project_response.json()
    #print(project_data)

    # Build the formatted output structure
    output = {
        "id": project_data['id'],
        "name": project_data['name'],
        "contributors": [{"name": contributor['name'], "email": contributor['email']} for contributor in contributors_data],
        "contents": {
            "meta": {
                "folders_count": sum(1 for item in tree_data if item['type'] == 'tree'),
                "files_count": sum(1 for item in tree_data if item['type'] == 'blob')
            },
            "folders": [
                {
                    "folder_name": folder['name'],
                    "folder_path": folder['path'],
                    "folder_mode": folder['mode']
                }
                for folder in tree_data
                if folder['type'] == 'tree'
            ],
            "files": [
                {
                    "file_name": file['name'],
                    "file_path": file['path'],
                    "file_mode": file['mode']
                }
                for file in tree_data
                if file['type'] == 'blob'
            ]
        }
    }

    # Generate the current timestamp
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    
    # Generate the file name with the current timestamp
    file_name = f"repo-query-{timestamp}.json"
    
    # Output the result
    print(json.dumps(output, indent=4))
 
    # Write the output to the file
    with open(file_name, 'w') as file:
        json.dump(output, file, indent=4)
    
        print(f"Output written to {file_name}")
except json.JSONDecodeError as e:
    print("Error decoding JSON response:" , e)

```
