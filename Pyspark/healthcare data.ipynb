```
from pyspark.sql import SparkSession, functions as F
 
spark=SparkSession.builder.getOrCreate()

path = '/FileStore/tables/Eligibility_Mock_Data___US_Healthcare_Bootcamp-7.xlsx'
dataFormat = "com.crealytics.spark.excel"
 
enrollment_sheet_name="Enrollment Mock Data"
demographics_sheet_name="Demographics Mock Data"
gender_crosswalk = "Gender Crosswalk"
group_crosswalk = "Group Crosswalk"
relationship_crosswalk = "Relationship Crosswalk"
coverage_crosswalk = "Coverage Crosswalk"
plan_crosswalk = "Plan Crosswalk"

enrollment_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{enrollment_sheet_name}'!").load(path)
demographics_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{demographics_sheet_name}'!").load(path)
gender_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{gender_crosswalk}'!").load(path)
group_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{group_crosswalk}'!").load(path)
relationship_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{relationship_crosswalk}'!").load(path)
coverage_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{coverage_crosswalk}'!").load(path)
coverage_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{coverage_crosswalk}'!").load(path)
plan_df = spark.read.format(dataFormat).option("inferschema", True).option("header", True).option("dataAddress", f"'{plan_crosswalk}'!").load(path)
```

```
final_df = enrollment_df.join(
    demographics_df,
    enrollment_df.member_id == demographics_df.member_id,
    "left"
).drop(demographics_df["member_id"])
```

```
final_df.display()
```

```
final_df = final_df.join(
    gender_df,
    final_df.gender == gender_df.code,
    "left"
) \
.drop("code") \
.withColumnRenamed("rollup_description", "gender_description") \
.withColumnRenamed("gender", "gender_code") \
.withColumnRenamed("rollup_code", "gender")

final_df = final_df.join(
    group_df,
    final_df.group_id == group_df.group_id,
    "left"
) \
.drop(group_df["group_id"])

final_df = final_df.join(
    coverage_df,
    final_df.coverage_type == coverage_df.coverage_id,
    "left"
) \
.drop(coverage_df["coverage_id"])

final_df = final_df.join(
    plan_df,
    final_df.plan_id == plan_df.plan_id,
    "left"
) \
.select(final_df["*"], plan_df["plan_name"], plan_df["benefit_type"], plan_df["effective_date"].alias("plan_effective_date"), plan_df["termination_date"].alias("plan_termination_date"))

final_df.display()

# .join(
#     relationship_df,
#     final_df.relationship == relationship_df.rollup_code,
#     "left"
# ) \
# .drop("rollup_code") \
# .withColumnRenamed("rollup_description", "relationship_description") \
```
