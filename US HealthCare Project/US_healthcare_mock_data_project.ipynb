#Imports 

```
from pyspark.sql import functions as F
```

#Mapping Gender Logic

```
gender_logic = demographic_data.alias("d").join(
    gender_crosswalk.alias("g"),
    F.col("d.gender") == F.col("g.code"),
    "inner" 
    ).drop(F.col("d.gender")
       ).select(F.col("d.*"),
         F.col("g.rollup_description").alias("gender"))
# demographic_data = demographic_data.join(gender_crosswalk, demographic_data["gender"] == gender_crosswalk["code"] ,'inner' )
```

```
display(gender_logic)
```

#Mapping Relationship Logic

```
relationship_logic = gender_logic.alias("g").join(
    relationship_crosswalk.alias("r"),
    F.col("g.relationship") == F.col("r.rollup_code"),
    "inner"
).drop(F.col("g.relationship")
       ).select(F.col("g.*"),
                F.col("r.*"))
```

```
display(relationship_logic)
```

#Mapping Enrollment Logic

```
enrollment_join = relationship_logic.alias("r").join(
    enrollment_data.alias("e"),
    "member_id",
    "left")
    
```

```
display(enrollment_join)
```

#Mapping Group Logic

```
group_logic = enrollment_join.alias("e").join(
    group_crosswalk.alias("g"),
    "group_id",
    "left"
)
```

```
display(group_logic)
```

#Mapping Coverage Logic

```
coverage_logic = group_logic.alias("g").join(
    coverage_crosswalk.alias("c"),
    F.col("g.coverage_type") == F.col("c.coverage_id"),
    "left"
).drop("coverage_type")
```

```
display(coverage_logic)
```

#Mapping Plan Logic

```
plan_logic = coverage_logic.alias("c").join(
    plan_crosswalk.alias("p"),
    "plan_id",
    "left"
)
```

```
display(plan_logic)
```

#Country,Souce_File_Name,Enrollment_Vendor_Name Logic

```
result_df = plan_logic.drop("race","ethnicity","member_status")
```

```
result_df = result_df.withColumnRenamed("vendor","Enrollment_Vendor_Name") \
    .withColumn("Country", F.lit("US")) \
    .withColumn("Souce_File_Name", F.lit("Member Enrollment"))
```

```
display(result_df)
```

#DOB Logic

```
result_df = result_df.withColumn("dob",F.date_format("dob","MM/dd/yyyy"))
```

```
display(result_df)
```

#Member Prefix and suffix logic

```
from pyspark.sql.types import IntegerType, StringType, BooleanType, DateType
result_df = result_df.withColumn("Member_Prefix_Name", 
                                F.when((result_df["gender"] == "Female") & 
                                      (result_df["rollup_description"].isin("Spouse", "Mother", "Grand Mother")),
                                      "Mrs."
                                     )
                                 .when(result_df["gender"] == "Male", "Mr.")
                                 .otherwise("Miss.")
                                ) \
                                .withColumn("Member_Suffix_Name", F.lit("None").cast(StringType()))
```

```
display(result_df)
```

#Record_id Logic

```
result_df = result_df.withColumn("Record_ID",F.monotonically_increasing_id())
```

#Abacus Member Id logic

8 Digit Unique ID 

Windows rownumber if member_id is same {1 Digit} + Year_of_birth{4 Digit} + Initials of Name {2 Digit} + Person code {1 Digit }

```
from pyspark.sql import Window
windows_spec = Window.partitionBy("member_id").orderBy("record_ID")
result_df = result_df.withColumn("row_num",F.row_number().over(windows_spec))
```

```
result_df = result_df.withColumn("Unique_id",
                                 F.concat(F.col("row_num").cast("string").substr(1,1),
                                          F.col("dob").cast("string").substr(1,4),
                                          F.col("first_name").substr(1,1),
                                          F.col("last_name").substr(1,1),
                                          F.col("person_code").cast("string").substr(1,1)))
```

```
display(result_df.select("row_num","dob","first_name","last_name","person_code","Unique_id"))
```

```
display(result_df.groupBy("Unique_id",).count().filter("count > 1"))
```

```
from pyspark.sql.types import IntegerType, StringType, BooleanType, DateType
result_df = result_df.withColumn("Member_Home_Phone", F.lit(0).cast(IntegerType())) \
                    .withColumn("Member_Work_Phone", F.lit(0).cast(IntegerType())) \
                    .withColumn("Member_Mobile_Phone", F.lit(0).cast(IntegerType())) \
                    .withColumn("Member_Email", F.lit("None").cast(StringType())) \
                    .withColumn("Member_Is_Deceased", F.lit(False).cast(BooleanType())) \
                    .withColumn("Member_Date_of_Death", F.lit(None).cast(DateType())) \
                    .withColumn("Member_Deceased_Reason", F.lit("None").cast(StringType())) \
                    .withColumn("Enrollment_SubGroup_ID", F.lit("None").cast(StringType())) \
                    .withColumn("Enrollment_SubGroup_Name", F.lit("None").cast(StringType())) \
                    .withColumn("Enrollment_Dental_Effective_Date", F.lit(None).cast(DateType())) \
                    .withColumn("Enrollment_Dental_Termination_Date", F.lit(None).cast(DateType())) \
                    .withColumn("Enrollment_Vision_Effective_Date", F.lit(None).cast(DateType())) \
                    .withColumn("Enrollment_Vision_Termination_Date", F.lit(None).cast(DateType())) \
                    .withColumn("File_Ingestion_Date", F.lit(None).cast(DateType()))


```

```
column_rename_dict = {
    "Record_ID": "Abacus_Record_ID",
    "Unique_id": "Abacus_Member_ID",
    "member_id": "Member_ID",
    "employee_id": "Subscriber_ID",
    "first_name": "Member_First_Name",
    "last_name": "Member_Last_Name",
    "middle_name": "Member_Middle_Name",
    "gender": "Member_Gender",
    "rollup_code": "Member_Relationship_Code",
    "rollup_description": "Member_Relationship_Description",
    "person_code": "Member_Person_Code",
    "address_1": "Member_Address_Line_1",
    "address_2": "Member_Address_Line_2",
    "city": "Member_City",
    "state": "Member_State",
    "county": "Member_County",
    "zip": "Member_Postal_Code",
    "Country": "Member_Country",
    "group_id": "Enrollment_Group_ID",
    "group_name": "Enrollment_Group_Name",
    "coverage_id": "Enrollment_Coverage_Code",
    "coverage_description": "Enrollment_Coverage_Description",
    "plan_id": "Enrollment_Plan_ID",
    "plan name": "Enrollment_Plan_Name",
    "benefit_type": "Enrollment_Plan_Coverage",
    "p_effective_date": "Enrollment_Medical_Effective_Date",
    "p_termination_date": "Enrollment_Medical_Termination_Date",
    "dob": "Member_Date_of_Birth"
}

for old_col, new_col in column_rename_dict.items():
    result_df = result_df.withColumnRenamed(old_col, new_col)
```
